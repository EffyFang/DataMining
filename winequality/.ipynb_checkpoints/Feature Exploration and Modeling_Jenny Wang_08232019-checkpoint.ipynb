## PCA

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import Normalizer
from sklearn.pipeline import make_pipeline

# Create scaler: scaler - standard works better here.
scaler = StandardScaler()
#scaler = Normalizer()

# Create PCA instance: model
pca = PCA()

# Create pipeline: pipeline
pipeline = make_pipeline(scaler, pca)

# Fit the pipeline to 'samples'
pipeline.fit(X)

#pca_result = pca.fit_transform(X)

# Plot the explained variances
features = range(1,pca.n_components_+1)
plt.bar(features, pca.explained_variance_)
plt.xlabel('PCA feature')
plt.ylabel('variance')
plt.xticks(features)
plt.show()
```

```python
npc = 8
pca = PCA(n_components=npc, copy=True)
X_standardized=scaler.fit_transform(X)
X_transformed = pca.fit_transform(X_standardized)
```

```python
pca_df = pd.DataFrame(columns = ['pca1','pca2','pca3','pca4','pca5','pca6','pca7','pca8'])

pca_df['pca1'] = X_transformed[:,0]
pca_df['pca2'] = X_transformed[:,1]
pca_df['pca3'] = X_transformed[:,2]
pca_df['pca4'] = X_transformed[:,3]
pca_df['pca5'] = X_transformed[:,4]
pca_df['pca6'] = X_transformed[:,5]
pca_df['pca7'] = X_transformed[:,6]
pca_df['pca8'] = X_transformed[:,7]

y=ro.round(pca.explained_variance_ratio_,3)

print ('Variance explained per principal component: {}'.format(y) )
```

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++',
                    max_iter = 400, n_init = 10, random_state = 0)
    kmeans.fit(X_standardized)
    wcss.append(kmeans.inertia_)
    
#Plotting the results onto a line graph to observe 'The elbow'
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Association')
plt.ylabel('WCSS') #within cluster sum of squares
plt.show()
```
