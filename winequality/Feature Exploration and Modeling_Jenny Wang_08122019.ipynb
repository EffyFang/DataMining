The 2nd, 3rd, and 4th important features that affect the alcohol quality the most for white and red wines are as the following: 

White: density (-0.31), chlorides (-0.21), volatile acidity (-0.19).

Red: volatile acidity (-0.39), sulphates (0.25), citric acid (0.23).

# Part 3. Modeling

Let's divide wine with quality score > 5 or not.

```python
y = white_wo_review.quality                  # set 'quality' as target
X = white_wo_review.drop('quality', axis=1)  # rest are features
print(y.shape, X.shape) 
```

Create a new y1 to record the records that quality score > 5

```python
y1 = (y > 5).astype(int)
y1.head(11)
```

draw a histogram to see the two groups' aggregated distribution:

```python
ax = y1.plot.hist(color='blue')
ax.set_title('Wine quality distribution', fontsize=14)
ax.set_xlabel('aggregated target value')
```

## Random Forest

```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, log_loss
from sklearn.metrics import confusion_matrix
```

```python
seed = 8 # set seed for reproducibility
X_train, X_test, y_train, y_test = train_test_split(X, y1, test_size=0.3, random_state=seed)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Instantiate the Random Forest Classifier
RF_clf = RandomForestClassifier(random_state=seed)
RF_clf
```

k-fold validation on training data and check the averaged scores:

```python
cv_scores = cross_val_score(RF_clf,X_train, y_train, cv=10, scoring='accuracy')
print('The accuracy scores for the iterations are {}'.format(cv_scores))
print('The mean accuracy score is {}'.format(cv_scores.mean()))
```
